```python
import torch
from agent import Agent, Policy

# 2. Instantiate policy and agent
policy = Policy(state_space=11, action_space=3)
agent  = Agent(policy)

# 3. Build a tiny “episode” using the real policy to produce log_probs
state = torch.zeros(11).numpy()  # start from an all-zero state

for t in range(3):
    # Let the policy choose an action (and return the correct log_prob)
    action, log_prob = agent.get_action(state)

    # Note: agent.get_action wraps policy.forward() similarly, but returns numpy→tensor conversions.
    # We can just call policy.forward() directly here to get a log_prob tied to parameters.

    # Store what we need for REINFORCE
    next_state = torch.zeros(11).numpy()  # we’ll pretend the next state is also zeros
    reward = 1.0                           # give a constant reward of 1.0 each step
    done = (t == 2)                        # end the episode at t=2

    # Append to the agent’s buffers (the same way store_outcome would)
    agent.states.append(torch.from_numpy(state).float())
    agent.next_states.append(torch.from_numpy(next_state).float())
    agent.action_log_probs.append(log_prob)        # this log_prob is connected to policy’s weights
    agent.rewards.append(torch.Tensor([reward]))
    agent.done.append(done)

    state = next_state  # move on (even though it’s still zeros)

# 4. Now call update_policy(); it should run without errors
agent.update_policy()

print("✅ Policy update ran successfully!")
# ----- END FIXED SANITY CHECK CELL -----
```
